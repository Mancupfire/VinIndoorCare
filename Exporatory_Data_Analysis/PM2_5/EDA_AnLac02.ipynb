## Evaluation

Correlation Matrix

```python
df_corr = pd.DataFrame({
    'Room 22': room22['PM2.5'].values,
    'Room 23': room23['PM2.5'].values,
    'Room 24': room24['PM2.5'].values,
    'General Room': room_gen['PM2.5'].values
})

corr_matrix = df_corr.corr()

print("Correlation matrix:")
print(corr_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of PM2.5 Sensors An Lac 2')
plt.show()
```

R-Squared Metrics

```python
r_squared_matrix = corr_matrix ** 2
print("R-squared Matrix:")
print(r_squared_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(r_squared_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title("R-squared Matrix of PM2.5 Across Sensors")
plt.show()
```

```python
def missing_data(data):
    total = data.isnull().sum()
    percent = (data.isnull().sum()/data.isnull().count()*100) # Tính tỉ lệ phần trăm của những cột bị trống
    tt = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    types = []
    for col in data.columns:
        dtype = str(data[col].dtype)
        types.append(dtype)
    tt['Types'] = types
    return(np.transpose(tt))
```
