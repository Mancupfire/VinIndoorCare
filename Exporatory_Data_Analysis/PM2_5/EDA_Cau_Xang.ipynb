```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
general = pd.read_csv(r"C:\Users\Administrator\Desktop\V_Indoor_Care\Data\General.csv")
room30 = pd.read_csv(r"C:\Users\Administrator\Desktop\V_Indoor_Care\Data\Room30.csv")
room31 = pd.read_csv(r"C:\Users\Administrator\Desktop\V_Indoor_Care\Data\Room31.csv")
room32 = pd.read_csv(r"C:\Users\Administrator\Desktop\V_Indoor_Care\Data\Room32.csv")
```

```python
print(general.isnull().sum())
print(room30.isnull().sum())
print(room31.isnull().sum())
print(room32.isnull().sum())
```

```python
general['PM2.5'].interpolate(method="linear", inplace=True)
room30['PM2.5'].interpolate(method="linear", inplace=True)
room31['PM2.5'].interpolate(method="linear", inplace=True)
room32['PM2.5'].interpolate(method="linear", inplace=True)
```

```python
print(general.isnull().sum())
print(room30.isnull().sum())
print(room31.isnull().sum())
print(room32.isnull().sum())
```

```python
gen = int(len(general['PM2.5']) * 0.25 ) 
r30 = int(len(room30['PM2.5']) * 0.25 ) 
r31 = int(len(room31['PM2.5']) * 0.25 ) 
r32 = int(len(room32['PM2.5']) * 0.25 ) 
```

```python
plt.figure(figsize=(14, 6))
plt.plot(general['PM2.5'][:gen], label='PM2.5', color = "red")
plt.plot(room30['PM2.5'][:r30], label='PM2.5', color = "green")
plt.plot(room31['PM2.5'][:r31], label='PM2.5', color = "blue")
plt.plot(room32['PM2.5'][:r32], label='PM2.5', color = "yellow")
plt.title('Compare PM2.5 Across Sensors at Cau Xang Primary School')
plt.xlabel('Time')
plt.ylabel('PM2.5')
plt.legend()
plt.grid(True)
plt.savefig('compare_pm25_across_sensors.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
data = pd.DataFrame({
    'General': general['PM2.5'].values,
    'Room 30': room30['PM2.5'].values,
    'Room 31': room31['PM2.5'].values,
    'Room 32': room32['PM2.5'].values
})
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1, square=True)
plt.title("Correlation Matrix of PM2.5 at Cau Xang Primary School")
plt.savefig('Corr_Matrix.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data.corr()**2, annot=True, cmap='YlGnBu', vmin=0, vmax=1, square=True)
plt.title("R-squared Matrix of PM2.5 at Cau Xang Primary School")
plt.savefig('R_Squared_matrixpng', dpi=300, bbox_inches='tight')
plt.show()
```
